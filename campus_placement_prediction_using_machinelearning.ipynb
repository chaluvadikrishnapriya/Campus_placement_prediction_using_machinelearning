```python
X = data.drop('status',axis=1)
y= data['status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)

svm = svm.SVC()
svm.fit(X_train,y_train)

knn=KNeighborsClassifier()
knn.fit(X_train,y_train)

dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)

rf=RandomForestClassifier()
rf.fit(X_train,y_train)

gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = svm.predict(X_test)
y_pred3 = knn.predict(X_test)
y_pred4 = dt.predict(X_test)
y_pred5 = rf.predict(X_test)
y_pred6 = gb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score1=accuracy_score(y_test,y_pred1)
score2=accuracy_score(y_test,y_pred2)
score3=accuracy_score(y_test,y_pred3)
score4=accuracy_score(y_test,y_pred4)
score5=accuracy_score(y_test,y_pred5)
score6=accuracy_score(y_test,y_pred6)
print(score1,score2,score3,score4,score5,score6)
```

```python
final_data = pd.DataFrame({'Models':['LR','SVM','KNN','DT','RF','GB'],
            'ACC':[score1*100,
                  score2*100,
                  score3*100,
                  score4*100,
                  score5*100,score6*100]})
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming y_test and y_pred1 to y_pred6 are already defined

# List of prediction arrays for each model
predictions = [y_pred1, y_pred2, y_pred3, y_pred4, y_pred5, y_pred6]

# Loop through the predictions and print the confusion matrix and accuracy for each
for i, y_pred in enumerate(predictions, start=1):
    # Compute accuracy score
    score = accuracy_score(y_test, y_pred)
    print(f"Accuracy for Model {i}: {score:.4f}")
    
    # Compute confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Print the confusion matrix
    print(f"Confusion Matrix for Model {i}:\n{cm}\n")
    
    # Plot the confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix for Model {i} (Accuracy: {score:.4f})")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```
